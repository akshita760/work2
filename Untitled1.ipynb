```python
import math
import csv
import random
```

```python
def load_csv(filename):
    dataset=[]
    with open(filename,'r') as file:
        reader=csv.reader(file)
        
        for row in reader:
            dataset.append(row)
        return dataset    
        
```

```python
def str_to_float(dataset):
    for row in dataset:
        for col in range(len(row)):
            
            row[col]=float(row[col])
    return dataset                     
```

```python
def maxmin(dataset):
    minmax=[]
    for row in range(len(dataset[0])):
        col=[col[row] for col in dataset]
        min_values=min(col)
        max_values=max(col)
        minmax.append([min_values,max_values])
    return minmax       
            
```

```python
def normalize_dataset(maxmin,dataset):
    numer=0
    denom=0
    for row in dataset:
        for col in range(len(row)):
            numer=row[col]-minmax[col][0]
            denom=minmax[col][1]-minmax[col][0]
            row[col]= numer/denom
            print(row[col])
            
```

```python
def cross_validation(dataset,n_folds):
    dataset_split=[]
    dataset_copy=list(dataset)
    fold_size=int(len(dataset)/n_folds)
    for i in range(len(n_folds)):
        fold=[]
        while fold<fold_size:
            index=random.randrange(len(dataset_copy))
            fold.append(dataset_copy.pop(index))
            dataset_split.append(fold)
    return dataset_split        
```

```python
def accuracy_metrics(actual, predicted):
    correct = 0
    for i in range(len(actual)):
        if actual[i] == predicted[i]:
            correct += 1
    return correct / len(actual) * 100

def predict(row, coef):
    yhat = coef[0]
    for i in range(len(row) - 1):
        yhat += coef[i + 1] * row[i]
    return 1.0 / (1.0 + math.exp(-yhat))

def evaluate_algorithm(dataset, algorithm, n_folds, learning_rate,nb_epoch):
    scores=[]
    folds = cross_validation(dataset, n_folds)
    for fold in folds:
        train=list(folds)
        train.remove(fold)
        train=sum(train,[])
        test=[]
        for row in fold:
            row_copy=list(row)
            row_copy[-1]=None
            test.append(row_copy)
        predicted = algorithm(train,test,learning_rate,nb_epoch)

        actual = [row[-1] for row in fold]
        score = accuracy_metrics(actual, predicted)
        scores.append(score)
    return scores
    
```
